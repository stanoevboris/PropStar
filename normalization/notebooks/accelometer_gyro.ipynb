```python
accelerometer = ['accX', 'accY', 'accZ']
gyro = ['gyroX', 'gyroY', 'gyroZ']
```

```python
norm = Normalize(denormalized_table=data, denormalized_table_name="original_table")
```

```python
norm.create_entity(source_entity_name="original_table", target_entity_name="accelerometer", columns=accelerometer,
                   distinct_values=True, retain_columns=['accelerometer_id'])
```

```python
norm.create_entity(source_entity_name="original_table", target_entity_name="gyro", columns=gyro,
                   distinct_values=True, retain_columns=['gyro_id'])
```
